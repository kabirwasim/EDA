```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('datasets.csv', encoding_errors='ignore')
```

```python
# null value and sum of null
data.isnull().sum()

# dropna delete all null values and inplace true save the result in orginal file
data.dropna(inplace=True)
```

```python

data.shape
```

```python
# duplicate records
data.duplicated().sum()

# inside square brackets gives duplicated data rows in db
data[data.duplicated()]

data.drop_duplicates(inplace=True)
```

```python
data.dtypes

# changing data types

data['id'] = data['id'].astype(object)

data['host_id'] = data['host_id'].astype(object)
```

```python

# Histogram plot for price to analyse

sns.histplot( data=data , x='price')

plt.title('Price Distribution')
```

```python
# identifying outlier in price distribution

sns.boxplot(data=data , x='price')
```

```python
df = data[data['price'] < 1500]

sns.boxplot( data= df , x='price')
```

```python

plt.figure(figsize=(8 ,5))
sns.histplot(data=df , x='price',bins=100)
plt.xlabel('price')
plt.ylabel('Frequency')
```
